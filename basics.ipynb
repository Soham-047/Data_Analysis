##Objects
Below, the Student is an Object, which consists of multiple fields, namely name and age.

```python
name = "Soham"
age = 20

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def print_details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

stud1 = Student(name, age)
stud2 = Student("Soham", 21)

class Stud:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def print_details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

s1 = Stud("Soham", 20)
s2 = Stud("Soham", 25)

stud1.print_details()
stud2.print_details()

print("__________")

s1.print_details()
s2.print_details()
```

OOPS


```python
# Inheritence + overriding

class A(object):
    def m1(self):
        print("This is class 1")

class B(A):
    def m2(self):
        print("This is class 2")

    def m1(self):
        print("This is class 1 but modified") # if this wasn't the case, i.e the initial function from class 1 was not overridden, then it would foloow the same emthod that is being defined in classs A

class C(B):
    def m3(self):
        print("This is class 3")

c = C()
c.m1()
c.m2()
c.m3()
```
